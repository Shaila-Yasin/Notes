# Python `re` Module: Regular Expressions Basics

Personal notes covering the core building blocks of Python's `re` module: the main functions (`search`, `findall`, `match`, `sub`, `split`), character classes, the dot operator, quantifiers, anchors, and character sets each with a short explanation followed by a runnable example.

## Core `re` functions

### `re.search()`  find the first match anywhere in a string
Scans the whole string and returns the **first** match it finds (or `None` if there's no match). `\D{3} t` here means: exactly 3 non-digit characters, followed by a space, followed by the letter `t`.

```python
import re
S =  "Welcom to Python programming"
res = re.search(r"\D{3} t", S)
print(res.group())
```

Another `re.search()` example looking for a fixed substring (`"AI"`) anywhere inside a sentence.

```python
text = " I Love AI Engineering"
result = re.search(r"AI", text)
print(result.group())
```

### `re.findall()`   get *all* matches, not just the first
Returns every non-overlapping match of the pattern as a list. Useful when the pattern can occur more than once in the string (unlike `search`, which stops at the first hit).

```python
text = "Python, Java Python, C++"
result = re.findall(r"Python", text)
print(result)
```

**Note:** `re.findall()` always returns a `list`  even with a single match, or an empty list `[]` when nothing matches.

```python
print(type(result))
```

### `re.match()` vs `re.search()`
- `re.match()` only checks for a match at the **very start** of the string. If the pattern isn't found at position 0, it returns `None` (and calling `.group()` on `None` would raise an error).
- `re.search()` checks the **entire** string, wherever the match occurs.

`re.IGNORECASE` (or `re.I`) makes the match case-insensitive, so `python` matches `Python` too.

```python
text1 = "Python programming"
text2 = "i like Python programing"
result1 = re.match(r"python", text1, re.IGNORECASE)
result2 = re.search(r"python", text2, re.IGNORECASE)
print(result1.group())
print(result2.group())
```

### `re.sub(pattern, replacement, string)`  find and replace
Replaces every match of `pattern` in `string` with `replacement`, and returns a **new** string (the original string is left unchanged).

```python
text3 = "Welcom to python programming"
result3 = re.sub(r"python", "Java", text3)
print(result3)
```

### `re.split(pattern, string)` split a string using a regex pattern
Works like `str.split()`, except the separator can be any regex pattern instead of just a fixed character much more flexible for messy or irregular text.

```python
text4 = "apple, banana, cherry"
result4 = re.split(",", text4)
print(result4)
```

Splitting on `\s` (a **single** whitespace character) each individual space becomes a split point, so consecutive spaces would leave empty strings in the result.

```python
text = "Python is a powerful programming"
print(re.split(r"\s", text))
```

Splitting on `\s+` (**one or more** whitespace characters) correctly handles multiple/irregular spaces by collapsing them into a single split point, generally the safer choice over `\s`.

```python
text = "Python  is a  powerful  programming"
print(re.split(r"\s+", text))
```

Splitting on `\d+` uses a run of one-or-more digits as the delimiter, so numbers of any length are removed and used as the split boundary.

```python
text = "apple123banana456orange"
print(re.split(r"\d+", text))
```

Splitting on a **character set** `[@#$]` , splits wherever *any single one* of these symbols appears.

```python
text = "Python@Java#C++$JavaScript"
print(re.split(r"[@#$]", text))
```

Splitting on `[,.?\s]+` (commas, periods, question marks, or whitespace one or more of any of them) to tokenize a sentence, then filtering out empty strings from the result with a list comprehension.

```python
text = "Hey, How are you doing?"
result = re.split(r"[,.?\s]+", text)
result = [word for word in result if word]
result
```

Splitting on `[^A-Za-z]+` a **negated character set**, meaning "anything that is NOT a letter". A quick way to pull just the words out of a sentence and drop punctuation/digits.

```python
text = "hello, My name is Zenn. I am an AI Engineer"
result = re.split(r"[^A-Za-z]+", text)
print(result)
```

### Capturing groups in `re.split()`
Wrapping the pattern in parentheses `()` turns it into a **capturing group**  the delimiters that matched are then *kept* in the output list instead of being discarded.

```python
text = "cat123dog456bird"
result1 = re.split(r"\d+", text)
result2 = re.split(r"(\d+)", text)
print(result1)
print(result2)
```

Splitting on `[-_]`  breaks apart a slug/identifier-style string on hyphens and underscores.

```python
text = "AI-Engineer_2026"
result = re.split(r"[-_]", text)
print(result)
```

## Character classes

### `\d` : matches a single digit (0–9)
Each digit is matched individually, so `"19"` produces two separate matches: `'1'` and `'9'`.

```python
text = 'my age is 19'
result = re.findall(r"\d", text)
print(result)
```

Using `\d+` (one or more) instead groups **consecutive** digits into a single match, giving `'19'` as one item instead of two.

```python
text = 'my age is 19'
result = re.findall(r"\d+", text)
print(result)
```

### `\D` : matches anything that is **not** a digit

```python
text = 'my age is 19'
result = re.findall(r"\D+", text)
result = [word.strip() for word in result if word]
print(result)
```

### `\w` : matches a "word character": letters (a-z, A-Z), digits (0-9), and underscore (`_`)

```python
text = 'my age is 19 _'
result = re.findall(r"\w+", text)
print(result)
```

### `\W` :  matches everything else (spaces, punctuation, symbols, etc.)  anything that is **not** a word character

```python
text = 'my! age is  19?'
result = re.findall(r"\W+", text)
print(result)
```

### `\s` :  matches whitespace: space, tab, or newline

```python
text = 'my age is  19'
result = re.findall(r"\s", text)
print(result)
```

### `\S` : matches anything that is **not** whitespace (opposite of `\s`)
Applying `\s` to a multi-line string also picks up the `\n` newline character, as shown below.

```python
text = '''my name is Zenn.
i'm studying Artificial intelligence.'''
result = re.findall(r"\s", text)
print(result)
```

Using `\S+` (one or more non whitespace characters) is a quick way to split text into words/tokens without calling `re.split()` at all.

```python
text = '''my name is Zenn.
i'm studying Artificial intelligence.'''
result = re.findall(r"\S+", text)
print(result)
```

### Dot Operator `.`
Matches **any single character**, except a newline (`\n`), by default.

```python
re.findall(r".", "ABC")
```

Because `.` does **not** match newline characters, `.+` stops at every `\n` : which is why this multi line string is returned as separate matches instead of one long one.

```python
text = '''my name is \n Zenn.
i'm studying Artificial intelligence.'''
result = re.findall(r".+", text)
print(result)
```

### Quantifier

Quantifiers tell how many times something should occur.

`*` :  0 or more occurrences of the preceding character. So `ab*` matches `a` on its own, `ab`, `abb`, and so on, but the `b` is always optional.

```python
text = ["a", "ab", "abb", "abc", "abba"]
for word in text:
    result = re.findall(r"ab*", word)
    print(result)
```

`{2,3}`  matches between 2 and 3 repetitions of the preceding character (greedy: it grabs as many as it can up to the max).

```python
text = "abc abbc abbb abca"
result = re.findall(r"ab{2,3}", text)
print(result)
```

Same `*` quantifier applied to a full sentence , note that `a` alone is still a valid match, since zero `b`s satisfies `ab*`.

```python
text = "abc abbc abbb abca"
result = re.findall(r"ab*", text)
print(result)
```

`+` :  1 or more occurrences of the preceding character (unlike `*`, at least one is required).

```python
text = "abc a abbc abbb abca"
result = re.findall(r"ab+", text)
print(result)
```

`?` :  0 or 1 occurrence.<br>
The character right before it becomes optional (it can appear 0 times or exactly 1 time), which is handy for matching both spelling variants like `color`/`colour`.

```python
text = "color colour"
result = re.findall(r"colou?r", text)
print(result)
```

`{n}` :  matches exactly `n` repetitions. Combined here with `\b` (word boundary, see below), `\b[A-Za-z]{4}\b` finds whole words that are **exactly 4 letters long**.

```python
text = 'zenn John Doe Jerry'
print(re.findall(r"\b[A-Za-z]{4}\b", text))
```

Without a word boundary, `findall` happily matches `cat` **inside** `category` too , it doesn't care about word edges, just the raw character sequence.

```python
text = "cat category"
print(re.findall(r"cat", text))
```

### `\b` : word boundary
Ensures a match is a **whole/standalone** token, not part of a longer word or number. `\b\d{3}\b` matches exactly 3 digit numbers, but skips `4567` because it's a 4 digit number, not a standalone 3 digit one.

```python
text = "123 4567 890"
print(re.findall(r"\b\d{3}\b", text))
```

Confirms `\b` behavior with a specific literal number: matches the standalone `123`, but not the `123` that's part of `1234`.

```python
text = "123 1234"
print(re.findall(r"\b123\b", text))
```

`{3,4}` : matches numbers that are 3 to 4 digits long, as whole tokens (skips the 5 digit `78906`).

```python
text = "123 4567 890 78906"
print(re.findall(r"\b\d{3,4}\b", text))
```

`{3,}` : matches 3 **or more** digits, with no upper limit, still anchored to whole numbers via `\b`.

```python
text = "123  67 4567 890 45678"
print(re.findall(r"\b\d{3,}\b", text))
```

### Anchors

`^` : anchors the match to the **beginning** of the string. Even though `"Python"` appears twice in the text, only the one at the very start counts.

```python
text = "Python is easy Python"
print(re.findall(r"^Python", text))
```

`$` : anchors the match to the **end** of the string.

```python
text = "my name is Zenn"
print(re.findall(r"Zenn$", text))
```

### Sets `[]`

`[abc]` : matches a single character that is `a`, `b`, **or** `c` (any one of the characters listed inside the brackets).

```python
re.findall(r"[abc]", "apple banana")
```

`[A-Z]` : a range shorthand for "any uppercase letter". Combined with `^`, `^[A-Z]` checks specifically whether the string **starts** with an uppercase letter.

```python
print(re.findall(r"[A-Z]", "Apple Banana"))
print(re.findall(r"^[A-Z]", "Apple Banana"))
```

Applying `[a-z]` (any lowercase letter) to different strings. it only ever matches lowercase letters, silently skipping digits, spaces, and uppercase letters.

```python
print(re.findall(r"[a-z]", "apple banana hii"))
print(re.findall(r"[a-z]", "Hi 234 like"))
```

`[0-9]` : the digit equivalent of `\d`, matches a single digit character.

```python
print(re.findall(r"[0-9]", "3 apple 4 banana"))
```

`[^abc]` : a **negated set**: the `^` right after the opening bracket flips the meaning to "anything except `a`, `b`, or `c`" (note this is a different use of `^` than the start of string anchor seen earlier).

```python
re.findall(r"[^abc]", "apple banana")
```

### `|` : alternation (OR)
Matches either `cat` **or** `dog`  the regex equivalent of a logical "or" between two patterns.

```python
re.findall(r"cat|dog", "cat dog bird")
```

Revisiting `\D+` (one or more non digits) grabs everything from the start of the string up to (but not including) the first digit. Since this sentence has no digits at all, the whole string matches.

```python
S =  "Welcome to Python programming"
res = re.search(r"\D+", S)
print(res.group())
```

Regular Python string slicing applied to the matched text `.group()` returns a plain string, so normal `[start:end]` slicing rules apply to it.

```python
print(res.group()[0:8])
```

Slicing a different portion of the same matched string.

```python
print(res.group()[8:17])
```
